```python
def trend_line_func(X, y, deg):
    
    y_pred= np.polyfit(X, y, deg)
    if deg == 2:
        y_arr= y_pred[0]*X**2 + y_pred[1]*X + y_pred[2]
        trend= y_pred[0] + y_pred[1] + y_pred[2]
    elif deg == 1:
        y_arr= y_pred[0]*X + y_pred[1]
        trend= y_pred[0] + y_pred[1]

    return [y_arr, trend]
```
