## 5. SHAP analysis

```python
regarr= range(18)
X_plot_df= pd.DataFrame([])
tmp_ind_vals= []
reg_shap_vals= []
mean_shap_vals= []

for r in tqdm(regarr):
    X_bkg_df= X_freq_tot.iloc[df_r[df_r.fire_freq == 0].groupby('reg_indx').get_group(r+1).index][X_freq_tot.columns].sample(n= 50, random_state= 67).values
    X_fires_df= X_freq_tot.iloc[df_r[(df_r.fire_freq == 1)].groupby('reg_indx').get_group(r+1).index][X_freq_tot.columns] 
    X_shap_df= pd.concat([X_fires_df, \
            X_freq_df.iloc[dftemp[(dftemp.fire_freq == 0)].groupby('reg_indx').get_group(r+1).index][X_freq_tot.columns].sample(n= int(len(X_fires_df)/3), random_state= 67)]).sample(frac= 1.0) #X_freq_tot.iloc[df_r[(df_r.fire_freq == 0)].groupby('reg_indx').get_group(r+1).index][X_freq_tot.columns]
    X_plot_df= X_plot_df.append(pd.concat([X_shap_df.reset_index().drop(columns= ['index']), pd.DataFrame({'reg_indx': (r+1)*np.ones(len(X_shap_df), dtype= int)})], axis= 1), ignore_index= True)
    def f(X):
        freq_arr= zipd_model_shap(mdn_mod= mdn_zipd, X= X)
        return freq_arr

    explainer= shap.KernelExplainer(f, X_bkg_df)
    shap_values= explainer.shap_values(X_shap_df.values, nsamples= 500)
    tmp_ind_vals.append(np.argsort(np.mean(abs(shap_values), axis= 0)))
    reg_shap_vals.append(shap_values)
    mean_shap_vals.append(np.sort(np.mean(abs(shap_values), axis= 0)))

X_plot_df.to_hdf('../sav_files/shap_files/fire_freq_data_%s.h5'%freq_id, key= 'df', mode= 'w')
pd.DataFrame(np.vstack(reg_shap_vals), columns= X_freq_tot.columns).to_hdf('../sav_files/shap_files/fire_freq_shap_%s.h5'%freq_id, key= 'df', mode= 'w')
```
