```python
X_sizes_train, X_sizes_val, y_sizes_train, y_sizes_val, fire_size_train, fire_size_test, X_sizes_test, y_sizes_test= fire_size_data(res= '12km', \
                            dropcols= drop_col_func(mod_type= 'normal', add_var_flag= True, add_var_list= ['Delta_T', 'Lightning']), \
                            start_month= 420, tot_test_months= 12, threshold= 4, scaled= True, tflag= True) #tflag= True; scaled= True, rh_flag= True

X_sizes_train_df= pd.concat([X_sizes_train, X_sizes_val], sort= False).reset_index().drop(columns=['index'])
X_sizes_tot= pd.concat([X_sizes_train_df, X_sizes_test], sort= False).reset_index().drop(columns=['index'])
fire_size_tot= pd.concat([fire_size_train, fire_size_test], sort= False).reset_index().drop(columns=['index'])

clim_fire_size_df= pd.read_hdf('../data/clim_fire_size_12km_w2020_data.h5')
clim_fire_freq_df= pd.read_hdf('../data/clim_fire_freq_12km_w2020_rescaled_data.h5')
```

```python
clim_fire_freq_df= clim_fire_freq_df.dropna().reset_index().drop(columns=['index'])

rb_frac= 7/3 #fraction of no fires to fires
df1= clim_fire_freq_df[clim_fire_freq_df['fire_freq']==1] #[fire_freq_train_df.reg_indx.isin([1, 2, 3, 4, 5])]
n2= rb_frac*len(df1)
df2= clim_fire_freq_df[clim_fire_freq_df['fire_freq']==0] 
dftemp= df2.copy(deep= True)
df2= df2.sample(n= int(n2), random_state= 99)
df_r= pd.concat([df1, df2], sort= False)
```

```python
def trend_line_func(X, y, deg):
    
    y_pred= np.polyfit(X, y, deg)
    if deg == 3:
        y_arr= y_pred[0]*X**3 + y_pred[1]*X**2 + y_pred[2]*X + y_pred[3]
        trend= y_pred[0] + y_pred[1] + y_pred[2] + y_pred[3]
    elif deg == 2:
        y_arr= y_pred[0]*X**2 + y_pred[1]*X + y_pred[2]
        trend= y_pred[0] + y_pred[1] + y_pred[2]
    elif deg == 1:
        y_arr= y_pred[0]*X + y_pred[1]
        trend= y_pred[0] + y_pred[1]

    return [y_arr, trend]
```
